```python
%matplotlib inline
import pandas as pd
import os
import seaborn as sns
from functools import reduce
sns.set()
pd.set_option("display.max_columns", 20)
pd.set_option("display.max_rows", 100)
```

```python
def pull_year(filename):
    return int(pd.read_excel(filename, header=0).columns[0].split()[0])
    
mcas_list = os.listdir("mcas")
mcas_dict = {pull_year("mcas/"+m): pd.read_excel("mcas/"+m, header=1, index_col=[0,2]) for m in mcas_list if m.endswith("xlsx")}
```

```python
df = pd.Panel(mcas_dict)
df
```

```python
df.loc[:,["Milton", "Quincy"],'P+A %'].plot(figsize=(10,5))
```
